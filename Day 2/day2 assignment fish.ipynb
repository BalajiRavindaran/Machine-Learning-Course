```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
path = "/gdrive/My Drive/ML:March2020/Assignments/data/"
raw_data = pd.read_csv(path+"Fish.csv")
raw_data.head(10)
```

```
raw_data['Species'].unique()
```

```
from sklearn.preprocessing import LabelEncoder
speciesEncoder=LabelEncoder()
raw_data['Species']=speciesEncoder.fit_transform(raw_data['Species'])
raw_data.head()
```

```
y=raw_data['Weight']
x=raw_data.drop('Weight',axis=1)
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=2)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x_train, y_train)
```

```
train_pred = model.predict(x_train)

test_pred = model.predict(x_test)

score = model.score(x_train, y_train)
print('Train F1 Score = {} %'.format(round(score,4)*100))
score = model.score(x_test, y_test)
print('Test F1 Score = {} %'.format(round(score,4)*100))
```

```
val=list(np.arange(0.5,12,0.001))
from sklearn.linear_model import  Ridge
model=Ridge()
from sklearn.model_selection import GridSearchCV
parameters={'alpha':val}#rigid
clf=GridSearchCV(model,parameters,n_jobs=-2)
clf.fit(x_train,y_train)
```

```
print(clf.best_score_)
```
