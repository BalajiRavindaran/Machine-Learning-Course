```
s1='send us your password'
s2='send us your review'
s3='review your password'
s4='review us'
s5='send your password'
s6='send us your account'
x=[s1,s2,s3,s4,s5,s6]
x
```

```
def P_t(text):
  pattern='[0-9]'
  data=[re.sub(pattern,'',i) for i in text]
  pattern1='_'
  data=[re.sub(pattern,'',i) for i in data]
  return data
```

```
data=P_t(x)
data
```

```
from sklearn.feature_extraction.text import CountVectorizer
vec=CountVectorizer(tokenizer=tok,stop_words=sw,ngram_range=(1,1))
```

```
Vector_x=vec.fit_transform(data)
print(Vector_x.toarray())
```

```
vec.get_feature_names()
```

```
from sklearn.feature_extraction.text import TfidfTransformer
TfId=TfidfTransformer()
```

```
Vector_x=TfId.fit_transform(Vector_x)
print(Vector_x.toarray())
```

```
y_train=[1,0,0,1,1,1]
```

```
from sklearn.naive_bayes import BernoulliNB
classifier1=BernoulliNB()
classifier1.fit(Vector_x,y_train)
```
