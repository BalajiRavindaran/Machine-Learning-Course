```
import pandas as pd
import numpy as np
```

```
path='/gdrive/My Drive/ML:March2020/data/'
data=pd.read_csv(path+"emails.csv")
data.head(10)
```

```
data.shape
```

```
x=data['text']
y=data['spam']
```

```
y
```

```
print("spam count:{},label is {}".format(sum(y==1),'1'))
print("not-spam count:{},label is {}".format(sum(y==0),'0'))
```

```
x=x.values.tolist()
type(x)
```

```
from nltk.corpus import stopwords

import nltk
import re
nltk.download('punkt')
nltk.download('stopwords')
import string
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer('\w+')
```

```
def tok(x):
  return (tokenizer.tokenize(x))
```

```
sw=stopwords.words('english')
sw
```

```
def process_text(text):
  pattern = '[0-9]'
  
  data = [re.sub(pattern,'' , i) for i in text]
  
  pattern1= '_'

  data = [re.sub(pattern1, '' , i) for i in data]

  return data
```

```
data=process_text(x)
data
```

```
from sklearn.feature_extraction.text import CountVectorizer
Vectorizer = CountVectorizer(analyzer='word', tokenizer=tok, stop_words=sw, ngram_range=(1, 1))
```

```
Vector_x= Vectorizer.fit_transform(data)
print(Vector_x.toarray())
```

```
Vector_x.shape
```

```
Vectorizer.get_feature_names()
```

```
print(Vector_x.toarray())
```

```
Vector_x.shape
```

```
from sklearn.feature_extraction.text import TfidfTransformer
Tfid = TfidfTransformer()
```

```
Vector_x = Tfid.fit_transform(Vector_x)
print(Vector_x.toarray())
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Vector_x,np.array(y), test_size=0.25, random_state=2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Bernoulli

```
from sklearn.naive_bayes import BernoulliNB
classifier1=BernoulliNB()
classifier1.fit(x_train,y_train)
```

```
y_pred=classifier1.predict(x_test)
```

```
print(np.array(y_pred.shape))
```

```
targets=['Not spam','Spam']
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred,target_names=targets))
```

## Gaussian

```
from sklearn.naive_bayes import GaussianNB
classifier2 = GaussianNB()
classifier2.fit(x_train.toarray(), y_train)

y_pred2= classifier2.predict(x_test.toarray())

print(np.array(y_pred2.shape))

targets = ['Not Spam', 'Spam']
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred2))
print(classification_report(y_test, y_pred2, target_names=targets))
```

## Complement

```
from sklearn.naive_bayes import ComplementNB
classifier3 = ComplementNB()
classifier3.fit(x_train.toarray(), y_train)

y_pred3= classifier2.predict(x_test.toarray())

targets = ['Not Spam', 'Spam']
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred3))
print(classification_report(y_test, y_pred3, target_names=targets))
```
