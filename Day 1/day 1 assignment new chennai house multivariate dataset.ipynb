```
path = "/gdrive/My Drive/ML:March2020/Assignments/data/"
data_train = pd.read_csv(path+"chennai_house_multivariate_train.csv")
data_test = pd.read_csv(path+"chennai_house_multivariate_test.csv")
data_train.head(10)
```

```
cols=data_train.shape[1]
print("columns={}".format(cols))
```

Normalization

```
data_train = (data_train - data_train.mean()) / ( data_train.max() - data_train.min())  
```

Split X and Y

```
Y=np.array(data_train.iloc[:,cols-1:cols])
X=np.array(data_train.iloc[:,:cols-1])
X.shape,Y.shape
```

```
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=2)
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()

new_model = model.fit(x_train, y_train)

train_pred = new_model.predict(x_train)

test_pred = new_model.predict(x_test)

score = new_model.score(x_train, y_train)
print('Train F1 Score = {} %'.format(round(score,4)*100))

score = new_model.score(x_test, y_test)
print('Test F1 Score = {} %'.format(round(score,4)*100))
```
