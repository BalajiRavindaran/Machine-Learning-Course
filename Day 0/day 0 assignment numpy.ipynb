vectorize-make a scalar function work on vectors

```
def foo(x):
    if x % 2 == 1:
        return x**2
    else:
        return x/2
print('x = 23 returns ', foo(23))
print('x = 25 returns ', foo(25))
```

```
foo_v = np.vectorize(foo, otypes=[float])
print('x = [23,14,45] returns ', foo_v([23,14,45]))
print('x = [[23,14,45], [1, 2, 3]] returns ', foo_v([[23,14,45], [1, 2, 3]]))
```

apply a function column wise or row wise

```
np.random.seed(100)
array_x = np.random.randint(1,10,size=[4,4])
array_x
```

```
def max_minus_min(x):
    return np.max(x) - np.min(x)
```

```
print('Row wise: ', np.apply_along_axis(max_minus_min, 1, arr=array_x))
```

```
print('Column wise: ', np.apply_along_axis(max_minus_min, 0, arr=array_x))
```
